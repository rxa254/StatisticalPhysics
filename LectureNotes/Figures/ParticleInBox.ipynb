```python
# Loading needed libraries and setting up preferences
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#import scipy.signal as sig
import scipy.constants as const
from mpl_toolkits.mplot3d import Axes3D

# Update the matplotlib configuration parameters:
plt.rcParams.update({'font.size': 20,
                     'font.family': 'serif',
                     'figure.figsize': (10, 8),
                     'axes.grid': True,
                     'grid.color': '#555555'})
```

## do the Plots

```python
# define the wavefunction
L = 1
def Psi(nx, ny, x, y):
    z = np.sin(nx*np.pi/L*x) * np.sin(ny*np.pi/L*y)
    return z

fig = plt.figure(figsize=(11,11))

#ax = Axes3D(fig)
ax = fig.gca(projection='3d')

bsize = 1
x = np.arange(0, bsize*L, 0.03)
y = np.arange(0, bsize*L, 0.03)
x, y = np.meshgrid(x, y)
z = Psi(2, 1, x, y)

ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='hot', alpha=0.53, linewidth=0.13)
#ax.contourf(x, y, z, zdir='z', offset=-2, cmap=plt.cm.hot)
cset = ax.contourf(x, y, z, zdir='z', offset=-2.0, cmap=plt.cm.hot)
cset = ax.contourf(x, y, z, zdir='x', offset=-0.3, cmap=plt.cm.hot)
cset = ax.contourf(x, y, z, zdir='y', offset=+1.3, cmap=plt.cm.hot)

ax.set_zlim(-2, 2)
#plt.grid(color='gray', alpha=0.15, linestyle='dashed', linewidth=2)
plt.ylabel(r'y', fontsize=18);
plt.xlabel(r'x', fontsize=18);
#plt.zlabel(r'$\Psi$', fontsize=18);
#plt.title('Output Noise of AD797 stage (changing R97)')
#plt.legend([str(Ts[0])+' K', str(Ts[1])+' K', str(Ts[2])+' K'], loc=1, fontsize=16)
plt.xlim((-0.3*bsize, 1.3*bsize))
plt.ylim((-0.3*bsize, 1.3*bsize))
plt.savefig("ParticleBoxPlot.pdf", bbox_inches='tight')
plt.show()
```
